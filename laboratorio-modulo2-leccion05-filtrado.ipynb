```python
# Carga el dataset de trabajo
netflix_final = pd.read_csv("datos/netflix_final.csv", sep = ",").reset_index(drop = True)
netflix_final.head()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_2021 = netflix_final["release_year"] == 2021
condicion_movie = netflix_final["type"] == "Movie"
netflix_final[condicion_2021 & condicion_movie]
```

```python
netflix_final[(netflix_final["release_year"] == 2021) & (netflix_final["type"] == "Movie")]
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
df_titulo_duracion = netflix_final[["Title", "duration"]]
df_titulo_duracion[df_titulo_duracion["duration"] == "90 min"]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
netflix_final[(netflix_final["country"] == "India") & (netflix_final["release_year"] > 2018 )]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
netflix_final[(netflix_final["rating"] == "TV-14") | (netflix_final["rating"] == "R")]
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
netflix_final[netflix_final["director"].str.contains("Christopher Nolan", case = True, na = False, regex = True)]
```

```python
pelis_scorsese = netflix_final[netflix_final["director"].str.contains("Martin Scorsese", case = True, na = False, regex = True)]
pelis_scorsese.shape
```

```python
df_directores = netflix_final[(netflix_final["director"] == "Christopher Nolan") | (netflix_final["director"] == "Martin Scorsese")]
df_directores.shape
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

10 pelis de esos países tienen más de un lenguaje

```python
pelis_paises = netflix_final[netflix_final["country"].isin(["United States", "United Kingdom", "Canada"])]

pelis_paises["Language"].value_counts()
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
titulos_fechas = netflix_final[netflix_final["release_year"].between(2015, 2020, inclusive = "both")]
titulos_fechas["Title"].unique()
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
titulos_puntuacion = netflix_final[netflix_final["IMDB Score"].between(6.5, 8.0, inclusive = "both")]
titulos_puntuacion[titulos_puntuacion["duration"].isnull()]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
netflix_final[(netflix_final["duration"] == "90 min") | (netflix_final["duration"] == "150 min" )]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
series = netflix_final[netflix_final["type"] == "TV Show"]
patron_regex = ".*Love,*"
series[series["Title"].str.contains(patron_regex, case= True, na = False, regex = True)]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron_regex1 = ".*Johnson.*"
netflix_final[netflix_final["director"].str.contains(patron_regex1, case= True, na = False, regex = True)]
```
